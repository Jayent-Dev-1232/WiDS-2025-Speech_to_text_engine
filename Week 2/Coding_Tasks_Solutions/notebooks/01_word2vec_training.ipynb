```python
# Importing necessary libraries
from gensim.models import Word2Vec
import pandas as pd
```

```python
# Reading the IMDB training data
train_df = pd.read_csv("D:\Projects\Winter_in_Data_Science\WiDS-2025-Speech_to_text_engine\Week 2\Coding_Tasks_Solutions\data\imdb_train.csv")
sentences = [review.split() for review in train_df['text']]
```

```python
# Using Word2Vec to train the model
w2v = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=2,
    sg=1
)
```

```python
# Saving the model
w2v.save("word2vec.model")
print(w2v.wv.most_similar("good"))
```
