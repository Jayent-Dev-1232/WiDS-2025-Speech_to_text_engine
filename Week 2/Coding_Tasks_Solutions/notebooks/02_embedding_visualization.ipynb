```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
```

```python
# Loading the trained Word2Vec model
model = Word2Vec.load("word2vec.model")
```

```python
# Extracting words and their corresponding vectors
words = list(model.wv.index_to_key)[:300]
vectors = np.array([model.wv[word] for word in words])
```

```python
# Performing dimensionality reduction using t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
reduced = tsne.fit_transform(vectors)
```

```python
# Plotting the embeddings
plt.figure(figsize=(10, 10))
plt.scatter(reduced[:, 0], reduced[:, 1])
for i, word in enumerate(words):
    plt.annotate(word, xy=(reduced[i, 0], reduced[i, 1]))

plt.title("t-SNE visualization of Word2Vec embeddings")
plt.show()
```
